```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.image import imread
import cv2
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import holoviews as hv
from holoviews import opts
hv.extension('bokeh')
import json

from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Activation, Dropout, Flatten, Dense, Conv2D, MaxPooling2D
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.utils import plot_model
from tensorflow.keras import backend
from sklearn.metrics import confusion_matrix, classification_report

# import shap
from operator import itemgetter
```

```python
# 이미지 데이터 경로
data_dir = '/content/gdrive/MyDrive/CNN_Abnormal_Detection/dataset/'
train_path = data_dir + 'train/'
test_path = data_dir + 'test/'
```

```python
plt.figure(figsize=(10,8))
ok = plt.imread(train_path + 'O/00001_11O_1.jpg')
plt.subplot(1, 2, 1)
plt.axis('off')
plt.title("O", weight='bold', size=20)
plt.imshow(ok,cmap='gray')

ng = plt.imread(train_path + 'X1/00001_11X_1.jpg')
plt.subplot(1, 2, 2)
plt.axis('off')
plt.title("X", weight='bold', size=20)
plt.imshow(ng,cmap='gray')

plt.show()
```

```python
image_gen = ImageDataGenerator(rescale=1/255,
                               validation_split=0.3)
```

```python
image_shape = (300,300,1) # 300 × 300、graysclaed (full-color : 3)
batch_size = 32

train_set = image_gen.flow_from_directory(train_path,
                                            target_size=image_shape[:2],
                                            color_mode="grayscale",
                                            classes={'O': 0, 'X2': 1},
                                            batch_size=batch_size,
                                            class_mode='binary',
                                            shuffle=True)

validation_set = image_gen.flow_from_directory(
                                train_path, # same directory as training data
                                target_size=image_shape[:2],
                                color_mode="grayscale",
                                classes={'O': 0, 'X2': 1},
                                batch_size=batch_size,
                                class_mode='binary',
                                shuffle=True) # set as validation data

test_set1 = image_gen.flow_from_directory(test_path,
                                           target_size=image_shape[:2],
                                           color_mode="grayscale",
                                           classes={'O': 0},
                                           batch_size=batch_size,
                                           class_mode='binary',
                                           shuffle=False)

test_set2 = image_gen.flow_from_directory(test_path,
                                           target_size=image_shape[:2],
                                           color_mode="grayscale",
                                           classes={'X1': 1},
                                           batch_size=batch_size,
                                           class_mode='binary',
                                           shuffle=False)

test_set3 = image_gen.flow_from_directory(test_path,
                                           target_size=image_shape[:2],
                                           color_mode="grayscale",
                                           classes={'X2': 1},
                                           batch_size=batch_size,
                                           class_mode='binary',
                                           shuffle=False)
```

```python
# CNN 신경망 생성
backend.clear_session()
model = Sequential()
model.add(Conv2D(filters=16, kernel_size=(7,7), strides=2, input_shape=image_shape, activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=2))
model.add(Conv2D(filters=32, kernel_size=(3,3), strides=1, input_shape=image_shape, activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=2))
model.add(Conv2D(filters=64, kernel_size=(3,3), strides=1, input_shape=image_shape, activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=2))
model.add(Flatten())
model.add(Dense(units=224, activation='sigmoid'))
model.add(Dropout(rate=0.2))
model.add(Dense(units=1, activation='sigmoid'))
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
model.summary()
```

```python
model_save_path = '/content/gdrive/MyDrive/CNN_Abnormal_Detection/Abnormal_Detection3.hdf5'
early_stop = EarlyStopping(monitor='val_loss',patience=2)
checkpoint = ModelCheckpoint(filepath=model_save_path, verbose=1, save_best_only=True, monitor='val_loss')
```

```python
# 모델 학습하기
n_epochs = 20
results = model.fit_generator(train_set, epochs=n_epochs, validation_data=validation_set, callbacks=[early_stop,checkpoint])
```

```python
model_history = { i:list(map(lambda x: float(x), j)) for i,j in results.history.items() }
with open('/content/gdrive/MyDrive/CNN_Abnormal_Detection/model_history3.json', 'w') as f:
    json.dump(model_history, f, indent=4)
```

**정상, 파손1, 파손2 테스트하기**

```python
# test_set1 테스트하기
pred_probability = model.predict_generator(test_set1)
predictions = pred_probability > 0.5

plt.figure(figsize=(10,6))
plt.title("Confusion Matrix", size=20, weight='bold')
sns.heatmap(
    confusion_matrix(test_set1.classes, predictions),
    annot=True,
    annot_kws={'size':14, 'weight':'bold'},
    fmt='d',
    xticklabels=['O', 'X1'],
    yticklabels=['O', 'X1'])
plt.tick_params(axis='both', labelsize=14)
plt.ylabel('Actual', size=14, weight='bold')
plt.xlabel('Predicted', size=14, weight='bold')
plt.show()
```
